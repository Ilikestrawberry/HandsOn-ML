PBR: PBR(주가 / 주당순자산) 1.5이하인지 아닌지

PER: PER(주가 / 주당순이익) 15이하인지 아닌지

Debt ratio: 부채비율 100% 이하 10점 / 175%~100% 7점 / 250%~175% 4점 /

ROA over20: ROA(순이익 / 자산) 20%가 넘는지 아닌지

PRR under15: PRR(시가총액 / 연구개발비) 15이하

Cash flow: 최근분기 유동비율(유동자산 / 유동부채) 100% 이상

PSR under1: PSR(주가 / 주당매출액) 1 이하인지 아닌지

3y Increse Profit: 지난 3년간 영업이익 지속 증가

3y Decrese Debt: 지난 3년간 부채비율 지속 감소

3y Positive Profit: 지난 3년간 순이익
